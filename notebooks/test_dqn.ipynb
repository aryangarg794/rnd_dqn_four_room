```python
import numpy as np
import torch 
import gymnasium as gym
import matplotlib.pyplot as plt

from copy import deepcopy
from dqn.train import train_eps_greedy
from dqn.model import DQN
from rnd_exploration.dataset import MovingSet
%load_ext autoreload
%autoreload 2
```

```python
env = gym.make('CartPole-v1')
val_env = deepcopy(env)

agent = DQN(env=env, val_env=val_env, use_cnn=False, decay_steps=10000, lr=5e-4)
```

```python

_, rewards = train_eps_greedy(
    agent, 
    env,
)
```

```python
plt.style.use('ggplot')   

plt.title('Cartpole DQN Test')
xs = np.arange(0, 200000, 5000)
plt.plot(xs, rewards, label='Val Score')
plt.legend()
plt.grid(True)
plt.ylabel('Cumm. Reward')
plt.xlabel('Timestep')
plt.show()
```
